```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import numpy as np
```

```python
data = {
    'titular': [
    "equipo gana campeonato",
    "nuevo telefono inteligente",
    "jugador anota gol decisivo",
    "empresa lanza nueva computadora",
    "equipo gana torneo"
    ],

    'clase': [
        "deportes",
        "tecnologia",
        "deportes",
        "tecnologia",
        "deportes"
    ]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Convertir el texto en una matriz de caracteristicas
vectorizador = CountVectorizer()
X = vectorizador.fit_transform(df['titular'])
print(X.toarray())
```

```python
#Establever la variable objetivo
y = df['clase']

#Entrenar el modelo de Naive Bayes
modelo = MultinomialNB()
modelo.fit(X, y)

#Predecir la clase de un nuevo correo
nuevo_correo = ["equipo gana torneo"]
correo_vectorizado = vectorizador.transform(nuevo_correo)
prediccion = modelo.predict(correo_vectorizado)
print(prediccion)
```
