```python
import pandas as pd
df = pd.read_csv('part1_employees.csv')

#mostrar la info tecnica del df
df.info ()
```

```python
#Obtén todas las filas del depto "Eng" que además tengan perf_score >= 0.85,
#pero muestra solo las columnas: emp_id, name, city, perf_score, salary_mxn.
df_eng = df.loc[
    (df['dept'] == 'Eng') & (df['perf_score'] >= 0.85),
    ['emp_id', 'name', 'city', 'perf_score', 'salary_mxn']
]

df_eng
```

```python
#Del DataFrame completo, obtén las filas 3 a 8 (incluyendo 3 y excluyendo 8),
# y solo las columnas 0, 1, 5 y 6 (por posición).
tablanueva = df.iloc[3:8, [0, 1, 5, 6]]
tablanueva
```

```python
#goog_1616851413 Calcula:Filtra empleados que estén en city CDMX o GDL, con warnings == 0 y absences <= 1.
#Muestra emp_id, name, city, warnings, absences.

df_city = df.query(
    "city in ['CDMX','GDL'] and warnings == 0 and absences <= 1"
)[['emp_id', 'name', 'city', 'warnings', 'absences']]

df_city
```

Calcula:


```python
#salario promedio (salary_mxn)
#perf promedio (perf_score)
#total de warnings (warnings)
resumen = df.agg({
    'salary_mxn': 'mean',
    'perf_score': 'mean',
    'warnings': 'sum'
})

resumen
```

Por cada dept, calcula:

```python
#número de empleados
#salario promedio
#perf promedio
#Ordena el resultado por salario promedio descendente.
dept_summary = (
    df.groupby('dept')
      .agg(
          empleados=('emp_id', 'count'),
          salario_promedio=('salary_mxn', 'mean'),
          perf_promedio=('perf_score', 'mean')
      )
      .sort_values(by='salario_promedio', ascending=False)
)

dept_summary
```
