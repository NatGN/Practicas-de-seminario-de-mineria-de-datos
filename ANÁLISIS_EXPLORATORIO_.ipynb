```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('streaming_platform.csv')
df.info()
```

```python
#Horas vistas por la semana vs numero de sesiones
plt.figure(figsize=(6,4))
plt.scatter(data=df,x='hours_watched_per_week',y='number_of_sessions')
plt.title('Horas vistas por la semana vs numero de sesiones')
plt.xlabel('Horas de visitas por semana')
plt.ylabel('Numero de sesiones')
plt.show()
```

```python
#Edad vs ultimo inicio de sesion
plt.figure(figsize=(6,4))
plt.scatter(data=df,x='age',y='days_since_last_login')
plt.title('Edad vs ultimo incio de sesion')
plt.xlabel('Edad')
plt.ylabel('Ultimo inicio de sesion')
plt.show()
```

```python
#Edad vs Horas vistas por semana
plt.figure(figsize=(6,4))
plt.scatter(data=df,x='age',y='hours_watched_per_week')
plt.title('Edad vs ultimo incio de sesion')
plt.xlabel('Edad')
plt.ylabel('Horas de visitas por sermana')
plt.show()
```
