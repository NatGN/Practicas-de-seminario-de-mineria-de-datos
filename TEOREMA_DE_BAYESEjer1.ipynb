```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import numpy as np
```

```python
data = {
    'reseña': [
    "el producto funciona excelente",
    "muy buena calidad",
    "el producto llego dañado",
    "muy mala calidad",
    "excelente producto y calidad"
    ],

    'clase': [
        "satisfecho",
        "satisfecho",
        "no satisfecho",
        "no satisfecho",
        "satisfecho"
    ]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Convertir el texto en una matriz de caracteristicas
vectorizador = CountVectorizer()
X = vectorizador.fit_transform(df['reseña'])
print(X.toarray())
```

```python
#Establever la variable objetivo
y = df['clase']

#Entrenar el modelo de Naive Bayes
modelo = MultinomialNB()
modelo.fit(X, y)

#Predecir la clase de un nuevo correo
nuevo_correo = ["el producto llego dañado"]
correo_vectorizado = vectorizador.transform(nuevo_correo)
prediccion = modelo.predict(correo_vectorizado)
print(prediccion)
```
