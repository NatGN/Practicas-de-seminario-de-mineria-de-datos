```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('datos_estudiantes_decision.csv')
df.info()
```

```python
df.head()
```

```python
#1) Exploración visual
#Horas_estudio vs calificaciones
plt.figure(figsize=(6,4))
plt.scatter(data=df,x='horas_estudio',y='calificacion')
plt.title('Horas de estudio vs calificaciones')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificacion')
plt.show()
```

```python
#Horas_sueno vs nivel_estres
plt.figure(figsize=(6,4))
plt.scatter(data=df,x='horas_sueno',y='nivel_estres')
plt.title('Horas de sueño vs nivel de estres')
plt.xlabel('Horas de sueño')
plt.ylabel('Nivel de estres')
plt.show()
```

```python
#Uso_redes vs calificacion
plt.figure(figsize=(6,4))
plt.scatter(data=df,x='uso_redes',y='calificacion')
plt.title('Uso de redes vs calificaciones')
plt.xlabel('Uso de redes')
plt.ylabel('Calificacion')
plt.show()
```

```python
#Pearson para las 3

pearson_estudio_calf = df['horas_estudio'].corr(df['calificacion'],method='pearson')
pearson_sueno_estres = df['horas_sueno'].corr(df['nivel_estres'],method='pearson')
pearson_redes_calf = df['uso_redes'].corr(df['calificacion'],method='pearson')

print("Pearson: \n")
print("Horas de estudio vs calificacón: ",pearson_estudio_calf)
print("Horas de sueño vs Nivel de estres: ",pearson_sueno_estres)
print("Uso de redes vs calificacón: ",pearson_redes_calf)
```

```python
#Spearman para las 3
spearman_estudio_calf = df['horas_estudio'].corr(df['calificacion'],method='spearman')
spearman_sueno_estres = df['horas_sueno'].corr(df['nivel_estres'],method='spearman')
spearman_redes_calf = df['uso_redes'].corr(df['calificacion'],method='spearman')

print("Pearson: \n")
print("Horas de estudio vs calificacón: ",spearman_estudio_calf)
print("Horas de sueño vs Nivel de estres: ",spearman_sueno_estres)
print("Uso de redes vs calificacón: ",spearman_redes_calf)
```
